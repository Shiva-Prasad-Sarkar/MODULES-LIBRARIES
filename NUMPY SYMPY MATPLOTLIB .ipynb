```python
#1
#function minima and maxima
import sympy as sps
x = sps.symbols('x')
f = (x**2 - 7*x + 6) / (x - 10)
f2 = sps.diff(f, x)
f3 = sps.diff(f2,x)
cp = sps.solve(f2, x)#give a list output with values
for i in cp:
  if i==0:
    print('NO MAX OR MIN VALUE')
    break
  else:
    if f3.subs(x,i)<0:
      max = f.subs(x,i)
      print('The maximum value is =',max,'at point',i)
    else:
      min = f.subs(x,i)
      print('The minimum value is =',min,'at point',i)
```

```python
import sympy as smp
import numpy as np
def trapezodial_rule(f,a,b,n):
  x=smp.symbols('x')
  h=(b-a)/n
  f=smp.lambdify(x ,f)  #1/2 h [f(a)+f(b)]+h[(y1+y2+y3+....)]
  sum = (f(a)+f(b))/2
  for i in range(1,n):
    x_i = a + i*h
    sum+= f(x_i)
  return(h*sum)
x=smp.symbols('x')
f= x**2+2*x+3
a=-5
b=5
n=100 #efficiency will increase with the increasement of n
print(smp.integrate(f,(x,a,b)).evalf())#normally integration
print(trapezodial_rule(f,a,b,n))#with trapezodial rule
```

```python
import sympy as smp
import numpy as np
def trapezoidal_rule(f, a, b, n):
    x = smp.symbols('x')
    expr = smp.lambdify(x, f)  # Convert SymPy expression to a numerical function. expr(x)
    h = (b - a) / n
    sum = 0.5 * (expr(a) + expr(b))  # Add the first and last terms
    for i in range(1, n):
        x_i = a + i * h
        sum += expr(x_i)
    return sum * h
# Example usage
x = smp.symbols('x')
hdx = smp.sin(x)  # Define the function to be integrated
a = 0  # Lower limit of integration
b = np.pi  # Upper limit of integration
n = 1000  # Number of subintervals
# Perform numerical integration using the Trapezoidal Rule
h = trapezoidal_rule(hdx, a, b, n)
display(smp.Integral(hdx, (x, a, smp.pi)))
print("Approximate value of the integral:")
display(h)
```
